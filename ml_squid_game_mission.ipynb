```python
import pandas as pd
import scipy.stats as stats

df = pd.read_csv('Titanic.csv')
df2 = pd.crosstab(df['Gender'], df['Survived'])

chi2, p, dof, ef = stats.chi2_contingency(df2)
print(round(chi2, 2))
```

```python
import pandas as pd

df = pd.read_csv('Titanic.csv')
col = ['Survived', 'Gender', 'SibSp', 'Parch', 'Fare']
df = df[col]

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df['Gender'] = encoder.fit_transform(df['Gender'])
df['Fare'] = df['Fare'].fillna(df[df['Gender']==1]['Fare'].mean())

from statsmodels.formula.api import ols
model = ols('Survived ~ Gender + SibSp + Parch + Fare', data=df).fit()
print(model.params['Parch'])
```
